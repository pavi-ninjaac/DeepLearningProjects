# Five Ways to load and show an image in DL

```
# import libraries
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from keras.preprocessing.image import load_img
import keras
```

```
#load file from local system
from google.colab import files
uploaded = files.upload()
```

```
filename = 'cat_dog.jpg'
```

## 1) Using IPython

```
# auto return the image 
from IPython.display import Image
Image(filename = filename , width =224, height = 224)
```

## 2)Using matplotlib

```
img = plt.imread(filename) 
plt.imshow(img)
```

## 3)Using the keras API

```
keras_img = load_img(filename,target_size=(224,224))

input_arr=keras.preprocessing.image.img_to_array(keras_img)
input_arr = np.array([input_arr])  # Convert single image to a batch.
```

```
plt.imshow(keras_img)
```

# Using the Opencv

```
cv_img = cv2.imread(filename)
cv2.imshow('The image',cv_img) #openin image through colab will crash the secsion but it will work fine in jupyter notebook
```

```
# here we are using the plt imshow to show the image
plt.imshow(cv_img)
```

#### The default color pattern in cv2 is BGR we need change it to RGB before displaying it

```
plt.imshow(cv2.cvtColor(cv_img,cv2.COLOR_BGR2RGB))
```
