```
filename = 'cat_dog.jpg'
#load the image and show it

from keras.preprocessing.image import load_img
import matplotlib.pyplot as plt
import numpy as np
import keras

keras_img = load_img(filename,target_size = (224,224))

#shoe the image
plt.imshow(keras_img)
```

## Using the google mobileNet to classify image

```
# import the libraries
from keras.applications.mobilenet import MobileNet
```

```
#initializing the mobilenet
mobile_net = MobileNet()
```

### Need some preprocesing before using the mobilnet
1)change the image to array </br>
2) reshape it..need 4 dimentions for mobile net (chabge using np.expand_dims) <br>
3) mobilenet.preprocess_input(final_image)

```
from keras.preprocessing import image
```

```
img_arr = image.img_to_array(keras_img)

final_img = np.expand_dims(img_arr,axis = 0 )

preprocesed_img = keras.applications.mobilenet.preprocess_input(final_img)
```

```
#wll have a large set of data which is unknow
prdiction = mobile_net.predict(preprocesed_img)
print(prdiction)
```

```
# so that we need decodeing from mobilenet
from keras.applications import imagenet_utils
result = imagenet_utils.decode_predictions(prdiction)
print(result)
```

# MobileNet version 2 -- more accurate than 1st version

```
#import libraries

import keras
from keras.applications.mobilenet_v2 import MobileNetV2
from keras.applications import imagenet_utils # for output prediction decoding
from keras.applications.mobilenet_v2 import preprocess_input # for preprocessing
```

```
# load and display image
plt.imshow(keras_img)
```

```
#preprocesss before predict

keras_img_arr  =  image.img_to_array(keras_img)

#reshape it
final_keras_img = np.expand_dims(keras_img_arr , axis = 0 )

#preprocess imput
pre_img = preprocess_input(final_keras_img)
```

```
#initialize the mobilenet

mobilenet_v2 = MobileNetV2()
```

```
prediction = mobilenet_v2.predict(pre_img)

# decode the prediction to result

result = imagenet_utils.decode_predictions(prediction)
print(result)
```
